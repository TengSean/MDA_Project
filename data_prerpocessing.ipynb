```python
def except_hanlder(file):
    multi_file=False
    with open(file, 'r') as f:
        file  =f.read()
    #     print(file)
        file = [ file for file in file.split('}{')]
    #     file[-1] = file[-1][:-1]
        for i, data in enumerate(file):
    #         print(data)
            if i != 0:
                data = '{' + data
            if i != len(file)-1:
                data = data + '}'
            file[i] = data
        abc = file
    #     print(json.loads(json.dumps(file).strip()))
    #         print(data)
    print('[INFO]: Does the file duplicate',end='? ')
    if abc[0] == abc[1]:
        print('Yes')
        abc = set(abc)
    else:
        multi_file = True
        print('No')
    return abc, multi_file
#     t = eval(abc[1])
    
#     return 
    # print(abc)
```

```python
root_dir = './data/ID_Multi'
date_list = os.listdir(root_dir)

json_raw_list = []
rewrite=False
c = 0
for date in date_list:
    sub_dir = os.path.join(root_dir, date)
    json_file_list = os.listdir(sub_dir)
    for json_file in json_file_list:
        open_file = os.path.join(sub_dir, json_file)
        with open(open_file, 'r') as f:
            try:
                data = json.load(f)
            except ValueError as e:  # includes simplejson.decoder.JSONDecodeError
                rewrite = True
                c+=1
                error_class = e.__class__.__name__ #取得錯誤類型
                detail = e.args[0]                 #取得詳細內容
                t = time.strftime("%Y-%m-%d %H:%M:%S", time.localtime()) 
                errMsg = f"{error_class:<10}: {detail:<70} {t}"
                handlingMsg = f"[INFO]: Get exception"
                print(errMsg)
                print(handlingMsg)
                data, multi_file = except_hanlder(open_file)
        if(rewrite):
            print('[INFO]: Now re-writting the json file....')
            fc = 1
            rewrite_file = str(fc) + '_' + json_file
            rewrite_file = os.path.join(sub_dir, rewrite_file)
            for d in data:
                with open('./'+str(fc)+'.json','w') as rf:
                    rf.write(d)
                json_raw_list.append(eval(d))
                
            print('[INFO]: Removing the old version....')
#             os.remove(open_file)
#             if os.path.exists(open_file):
            print(f'[INFO]: {open_file} has been removed successfully!\n')
            rewrite = False
        else:
            json_raw_list.append(data)
print(f'Total {c} files have been rewrite!')
```

```python

# ===========================================================================================

# 砍除
def map1(x):
    if '[公告]' in x[0] and '砍除' in x[0] and 'Re:' not in x[0] and 'Fw:' not in x[0] and '停權' not in x[0]:#帳號砍除
        return True
    else:
        return False

def map1_proc(x):
    tmp = re.sub(r'[\W_]+', ' ', x[0])
    tmp = re.findall(r'[A-Za-z0-9]+', tmp)
#     tmp = [duplicate_user(t) for t in tmp]
#     tmp = list(filter(lambda x: len(x) > 0, tmp))
    tmp = [ [t, ['砍除', x[2]]] for t in tmp]
    return tmp
# ===========================================================================================


# 停權
def map2(x):
    if '[公告]' in x[0] and 'Re' not in x[0] and 'Fw:' not in x[0] and '停權' in x[0] and '砍除' not in x[0] and '砍帳' not in x[0]:
        return True
    else:
        return False

    
def map2_proc(x):
    tmp = re.sub(r'[\W_]+', ' ', x[0])
    tmp = re.findall(r'[A-Za-z0-9]+', tmp)
#     tmp = [duplicate_user(t) for t in tmp]
#     tmp = list(filter(lambda x: len(x) > 0, tmp))
    tmp = [ [t, ['停權', x[2]]] for t in tmp]
    return tmp
# ===========================================================================================
# 砍除&停權
def map3(x):
    if '[公告]' in x[0] and 'Re' not in x[0] and 'Fw:' not in x[0] and '停權' in x[0] and '砍除' in x[0] and '砍帳' not in x[0]:
        return True
    else:
        return False

    
def map3_proc(x):
    tmp = x[0].split('/')
    if '砍除' in tmp[0]:
        tmp1 = re.sub(r'[\W_]+', ' ', tmp[0])
        tmp1 = re.findall(r'[A-Za-z0-9]+', tmp1)
        tmp2 = re.sub(r'[\W_]+', ' ', tmp[1])
        tmp2 = re.findall(r'[A-Za-z0-9]+', tmp2)
        tmp = [ [[t, ['砍除', x[2]]] for t in tmp1], [[t, ['停權', x[2]]] for t in tmp2]  ]
    else:
        tmp1 = re.sub(r'[\W_]+', ' ', tmp[0])
        tmp1 = re.findall(r'[A-Za-z0-9]+', tmp1)
        tmp2 = re.sub(r'[\W_]+', ' ', tmp[1])
        tmp2 = re.findall(r'[A-Za-z0-9]+', tmp2)
        tmp = [ [[t, ['砍除', x[2]]] for t in tmp2], [ [t, ['停權', x[2]]] for t in tmp1] ]

    return tmp


# ===========================================================================================

# 不受歡迎
def map4(x):
    if '[公告]' in x[0] and 'Re' not in x[0] and 'Fw:' not in x[0] and '不受歡迎' in x[0] and '規範' not in x[0]:
        return True
    else:
        return False

    
def map4_proc(x):
    tmp = re.sub(r'[\W_]+', ' ', x[0])
    tmp = re.findall(r'[A-Za-z0-9]+', tmp)
    tmp = [ [t, ['不受歡迎', x[2]]] for t in tmp]
    return tmp
# ===========================================================================================

def map5(x):
    if '[公告]' in x[0] and 'Re' not in x[0] and 'Fw:' not in x[0] and'機器人' in x[0]:
        return True
    else:
        return False
    
    
def map5_proc(x):
    start = '砍除帳號如下：'
    end = '帳號站長'
    st = x[1].find(start)
    en = x[1].find(end)
    tmp = x[1][st:en]
    tmp = re.sub(r'[\W_]+', ' ', tmp)
    tmp = re.findall(r'[A-Za-z0-9]+', tmp)
    tmp = [t for t in tmp if len(t) > 2]
    tmp = [ [t, ['機器人', x[2]]] for t in tmp]
    return tmp
# ***************===========================================================================================

# 超過五名帳號
def map6(x):
    if '超過五名帳號' in x[0] and '公告' in x[0] and 'Re:' not in x[0] and 'Fw:' not in x[0]:
        return True
    else:
        return False

def map6_proc(x):
    start = ['同一人或同一集團持有或使用：', '兩案涉案帳號如下：']
    end = '依多重帳號管理辦法'
    st = [x[1].find(s) for s in start]
    st = max(list(filter(lambda x: x>0, st)))
    en = x[1].find(end)
    tmp = x[1][st:en]
    tmp = re.sub(r'[\W_]+', ' ', tmp)
    tmp = re.findall(r'[A-Za-z0-9]+', tmp)
    tmp = [ [t, ['清除共同帳號', x[2]]] for t in tmp]
#     tmp = [t for t in tmp if len(t) > 2]
#     tmp = [ [t, ['機器人', x[2]]] for t in tmp]
    return tmp


# 清除帳號 24
def map0(x):
    if '[公告]' in x[0] and '清除' in x[0] and 'Re:' not in x[0] and 'Fw:' not in x[0] and '機器人' not in x[0] and '出售' not in x[0]:#清除大量異常帳號
        return True
    else:
        return False


# 清除共同帳號 13
def map01_proc(x):
    start = '下列帳號予以清除：'
    end = ['註： ', '上列部分帳號曾使用信箱認證。', '帳號站長']
    st = x[1].find(start)
    en = [ x[1].find(e) for e in end ]
    en = min(list(filter(lambda x: x >= 0, en)))
    tmp = x[1][st:en]
#     date = handle_date(x[2])
    # 有備註的
    if '註：' in tmp:
        tmp = tmp[:tmp.find('註：')]
        tmp = re.sub(r'[\W_]+', ' ', tmp)
        tmp = re.findall(r'[A-Za-z0-9]+', tmp)
        tmp = [ [t, ['清除共同帳號', x[2]]] for t in tmp]
        
#         tmp = [duplicate_user(t) for t in tmp]
#         tmp = list(filter(lambda x: len(x) > 0, tmp))
#         return [x[1][st:x[1].find('註：')],tmp]
#         tmp = [ '清除共同帳號', tmp , x[2]]
#         tmp.extend(date)
        return tmp
    # 沒有備註的
    else:
        tmp = re.sub(r'[\W_]+', ' ', tmp)
        tmp = re.findall(r'[A-Za-z0-9]+', tmp)
        tmp = [ [t, ['清除共同帳號', x[2]]] for t in tmp]
        
#         tmp = [duplicate_user(t) for t in tmp]
#         tmp = list(filter(lambda x: len(x) > 0, tmp))
#         return [x[1][st:en],tmp]
#         tmp = [ '清除共同帳號', tmp , x[2]]
#         tmp.extend(date)
        return tmp

# 清除跳板帳號 7
def map02_proc(x):
    start = '砍除帳號：'
    end = ['註：曾因帳號遭盜用報案者', '上列部分帳號曾使用信箱認證。']
    st = x[1].find(start)
    en = [ x[1].find(e) for e in end ]
    en = min(list(filter(lambda x: x > 0, en)))
    tmp = x[1][st:en]
#     date = handle_date(x[2])
    if '刪除ID:' in tmp:
        tmp = tmp.split()
        tmp = [tmp[i+1] for i,v in enumerate(tmp) if v == '刪除ID:']
        tmp = [ [t, ['清除跳板帳號', x[2]]] for t in tmp]
        
#         tmp = [duplicate_user(t) for t in tmp]
#         tmp = list(filter(lambda x: len(x) > 0, tmp))
#         tmp = ['清除跳板帳號', tmp, x[2]]
#         tmp.extend(date)
        return tmp
    else:
        tmp = re.sub(r'[\W_]+', ' ', tmp)
        tmp = re.findall(r'[A-Za-z0-9]+', tmp)
        tmp = [ [t, ['清除跳板帳號', x[2]]] for t in tmp]
        
#         tmp = [duplicate_user(t) for t in tmp]
#         tmp = list(filter(lambda x: len(x) > 0, tmp))
#         tmp = ['清除跳板帳號', tmp, x[2]]
#         tmp.extend(date)
        return tmp
    
# 清除VPN帳號 2
def map03_proc(x):
    start = [ '清除帳號如下：', '下列帳號予以刪除：']
    end = '帳號站長'
    st = [ x[1].find(s) for s in start ]
    st = min(list(filter(lambda x: x > 0, st)))
    en = x[1].find(end)
    tmp = x[1][st:en]
#     date = handle_date(x[2])
    tmp = re.sub(r'[\W_]+', ' ', tmp)
    tmp = re.findall(r'[A-Za-z0-9]+', tmp)
    tmp = [ [t, ['清除VPN帳號', x[2]]] for t in tmp]
    
#     tmp = [duplicate_user(t) for t in tmp]
#     tmp = list(filter(lambda x: len(x) > 0, tmp))
#     tmp = [ '清除VPN帳號' , tmp, x[2]]
#     tmp.extend(date)
    return tmp

# 不實資料 1
def map04_proc(x):
    start = '全數砍除.'
    end = '日後若再發現同一人使用或持有之帳號'
    st = x[1].find(start)
    en = x[1].find(end)
    tmp = x[1][st:en]
#     date = handle_date(x[2])
    tmp = re.sub(r'[\W_]+', ' ', tmp)
    tmp = re.findall(r'[A-Za-z0-9]+', tmp)
    tmp = [ [t, ['清除不實資料帳號', x[2]]] for t in tmp]
    
#     tmp = [duplicate_user(t) for t in tmp]
#     tmp = list(filter(lambda x: len(x) > 0, tmp))
#     tmp = [ '清除不實資料帳號', tmp, x[2]]
#     tmp.extend(date)
    return tmp

def f(x):
    prior_level = {'清除共同帳號':0, '清除跳板帳號':0, '清除VPN帳號':0, '清除不實資料帳號':0,
                   '機器人':     1,
                   '砍除':       2,
                   '停權':       3,
                   '不受歡迎':   4}
    mini = 100
    mini_idx = -1
    if len(x[1]) <= 2:
        return x
    for idx in range(0, len(x[1]), 2):
        if 'fake' in x[1][idx]:
            return [x[0], ['del', 'del']]
        key = x[1][idx]
        if prior_level[key] < mini:
            mini = prior_level[key]
            mini_idx = idx
    return [x[0], [x[1][mini_idx], x[1][mini_idx+1]]]

#     for idx, duplicate in enumerate(x[1])
def restore(x):
    type_date = x[0].split('/')
    return [type_date[0], x[1], type_date[1]]
    
def handle_date(x):
    mouth = {'Jan':'01', 'Feb':'02', 'Mar':'03', 'Apr':'04', 'May':'05', 'Jun':'06', 'Jul':'07', 'Aug':'08',
             'Sep':'09', 'Oct':'10', 'Nov':'11', 'Dec':'12'}
    day = {'Mon':'1', 'Tue':'2', 'Wed':'3', 'Thu':'4', 'Fri':'5', 'Sat':'6', 'Sun':'7'}    
    tmp = re.sub(r'[\W_]+', ' ', x).split(' ')
    return tmp[6]+'-'+mouth[tmp[1]]+'-'+tmp[2]+'-'+day[tmp[0]]

# 清除超過五名帳號
jj = json_rdd.filter(lambda x: map6(x))\
                .flatMap(lambda x: map6_proc(x))



# 清除共同
aa = json_rdd.filter(lambda x: map0(x) )\
                .filter(lambda x: '不實資料註冊' not in x[1])\
                .filter(lambda x: 'VPN' not in x[1])\
                .filter(lambda x: '跳板' not in x[1])\
                .filter(lambda x: '下列帳號予以清除：' in x[1])\
                .flatMap(lambda x: map01_proc(x))\
#                 .map(lambda x: [x[0], x[1]])\#==================
#                 .reduceByKey(add)\
#                 .map(lambda x: len(set(x[1])))#==================
#                 .map(lambda x: [ x[0], list(set(x[1]))])\


# # 清除跳板
bb = json_rdd.filter(lambda x: map0(x) )\
                .filter(lambda x: '不實資料註冊' not in x[1])\
                .filter(lambda x: 'VPN' not in x[1])\
                .filter(lambda x: '跳板' in x[1])\
                .flatMap(lambda x: map02_proc(x))\
#                 .map(lambda x: [x[0], x[1]])\#==================
#                 .reduceByKey(add)\
#                 .map(lambda x: len(set(x[1])))#==================
#                 .reduceByKey(add)\
#                 .map(lambda x: [ x[0], list(set(x[1]))])\

# # 清除vpn
cc = json_rdd.filter(lambda x: map0(x) )\
                .filter(lambda x: '不實資料註冊' not in x[1])\
                .filter(lambda x: 'VPN' in x[1])\
                .flatMap(lambda x: map03_proc(x))\
#                 .map(lambda x: [x[0], x[1]])\#==================
#                 .reduceByKey(add)\
#                 .map(lambda x: len(set(x[1])))#==================
# #                 .reduceByKey(add)\
# #                 .map(lambda x: [ x[0], list(set(x[1]))])\

# # 清除不實資料
dd = json_rdd.filter(lambda x: map0(x) )\
                .filter(lambda x: '不實資料註冊' in x[1])\
                .flatMap(lambda x: map04_proc(x))\
#                 .map(lambda x: [x[0], x[1]])\#==================
#                 .reduceByKey(add)\
#                 .map(lambda x: len(set(x[1])))#==================
#                 .reduceByKey(add)\
#                 .map(lambda x: [ x[0], list(set(x[1]))])\


# # 砍除 
ee = json_rdd.filter(lambda x: map1(x) )\
                .flatMap(lambda x: map1_proc(x))\
#                 .map(lambda x: [x[0], x[1]])\#==================
#                 .reduceByKey(add)\
#                 .map(lambda x: len(set(x[1])))#==================
#                 .reduceByKey(add)

# # # 停權 
ff = json_rdd.filter(lambda x: map2(x) )\
                .flatMap(lambda x: map2_proc(x))\
#                 .map(lambda x: [x[0], x[1]])\#==================
#                 .reduceByKey(add)\
#                 .map(lambda x: len(set(x[1])))#==================
# #                 .reduceByKey(add)


# # 砍除&停權 
gg = json_rdd.filter(lambda x: map3(x) )\
                .flatMap(lambda x: map3_proc(x))\
                .flatMap(lambda x: x)\
#                 .map(lambda x: [x[0], x[1]])\#==================
#                 .reduceByKey(add)\
#                 .map(lambda x: len(set(x[1])))#==================
# #                 .reduceByKey(add)

# 不受歡迎
hh = json_rdd.filter(lambda x: map4(x) )\
                .flatMap(lambda x: map4_proc(x))\
#                 .flatMap(lambda x: map4_proc(x))\
#                 .map(lambda x: [x[0], x[1]])\#==================
#                 .reduceByKey(add)\
#                 .map(lambda x: len(set(x[1])))#==================
#                 .reduceByKey(add)

# 機器人
ii = json_rdd.filter(lambda x: map5(x) )\
                .flatMap(lambda x: map5_proc(x))\
#                 .map(lambda x: [x[0], x[1]])\#==================
#                 .reduceByKey(add)\
#                 .map(lambda x: len(set(x[1])))#==================
#                 .reduceByKey(add)
a = ['rockx80', 'jeff2004618', 'alexyu33', 'ssjik', 'ccu15']

wrong_people = [['taboo00',['fake','xx']]]
# wrong_people = (['WWWWW123',['fake','xx']],  ['lllxxxlll',['fake','xx']], ['taboo00',['fake','xx']])
wrong_people = sc.parallelize(wrong_people)
                                            
clean_total = sc.union([aa, bb, cc, dd, ee, ff, gg, hh, ii, jj, wrong_people])\
                    .reduceByKey(add)\
                    .map(lambda x: f(x))\
                    .filter(lambda x: x[1][0] != 'del')

# clean_total.collect()
clean_total = clean_total.map(lambda x: [ x[1][0], [x[0], x[1][1]] ])\
                            .groupByKey()\
                            .map(lambda x: [x[0], list(x[1])])\
                            .flatMap(lambda x: [ [x[0], xx] for xx in x[1] ])\
                            .map(lambda x: [x[0]+'/'+x[1][1], x[1][0]])\
                            .groupByKey()\
                            .map(lambda x: [x[0], list(x[1])])\
                            .map(lambda x: restore(x))\
                            .map(lambda x: ( x[0], x[1], handle_date(x[2]), len(x[1])))\
#                             .sortByKey()

# clean_total.collect()
                                              
df = clean_total.toDF(['Type', 'ID', 'Date', 'ID_Count']).toPandas()
# wrong_people.collect()

# 不受歡迎 冤案 WWWWW123 lllxxxlll  taboo00
# rockx80, jeff2004618, alexyu33, ssjik, ccu15 同一人或同一集團持有或使用：, 依多重帳號管理辦法, 兩案涉案帳號如下：
```
